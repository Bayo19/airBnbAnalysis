```python
import pandas as pd
import numpy as np

dataset = pd.read_csv("listingss.csv")
```

```python
# Borough with most airBnb properties
borough_with_most_properties = (dataset
                                    .groupby(["neighbourhood"])
                                    .size()
                                    .sort_values(ascending=False)
                                    )
borough_with_most_properties
```

```python
# Borough with most reviews
borough_with_most_reviews = (dataset
                                .groupby(["neighbourhood"])[["neighbourhood", "number_of_reviews"]]
                                .sum()
                                .sort_values(by="number_of_reviews", ascending=False))
borough_with_most_reviews
```

```python
# Finding airBnB host with most properties
uk_host_most_properties_id = (dataset
                                    .groupby(["host_id", "host_name"])
                                    .size()
                                    .sort_values(ascending=False)
                                    .reset_index(name="number_of_properties"))
uk_host_most_properties_id = uk_host_most_properties_id.loc[uk_host_most_properties_id["number_of_properties"].idxmax()]["host_id"]

uk_host_most_properties = dataset.loc[dataset["host_id"] == uk_host_most_properties_id]
uk_host_most_properties
```

```python
# All boroughs where top host has 10 or more properties
properties_per_borough_top_host = uk_host_most_properties.groupby(["neighbourhood"]).size().sort_values(ascending=False)
props_per_borough_more_than_10 = pd.Series({k:v for k,v in properties_per_borough_top_host.to_dict().items() if v >= 10})
props_per_borough_more_than_10
```
